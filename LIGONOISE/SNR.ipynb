```python
import numpy as np
from matplotlib import pyplot as plt
Signal = np.loadtxt("LigoSignal.txt")
Signal[:,1] = Signal[:,1]*10**-21
Noise = np.loadtxt("LigoNoise.dat")
```

```python
SignalL = np.loadtxt("LivingstonLigoSignal.txt")
SignalL[:,1] = Signal[:, 1]*10**-21
plt.plot(Signal[:,0], Signal[:,1])
```

```python
# Get the DFT of the signal
# Hanford Data
T = Signal[-1,0]- Signal[0,0]
Signalhat = np.fft.fft(Signal[:,1])*T/len(Signal[:,1])
freqs = np.arange(len(Signalhat))/T
plt.figure(figsize=(10,10))
plt.plot(Signal[:,0], Signal[:,1])
plt.axhline(y=0, color="r")

```

```python
plt.figure(figsize=(10,10))
plt.plot(freqs[0:100], np.log(abs(Signalhat[0:100]))/np.log(10), "g", label = "signal")
plt.plot(Noise[0:2400,0], np.log(np.sqrt(Noise[0:2400,1]))/np.log(10),"r", label = "noise")
```
